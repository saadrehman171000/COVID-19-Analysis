```python
# Now Both of our datasets are ready to work with now we will start creating a clustering model 
covid_data.drop('Province/State', axis=1, inplace=True)
```

```python
from sklearn.cluster import KMeans

kmeans_model_1 = KMeans(n_clusters=3)
covid_data['cluster'] = kmeans_model_1.fit_predict(covid_data)
```
